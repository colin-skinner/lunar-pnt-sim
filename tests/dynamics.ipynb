```python
import sys
sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt
from lunanav.constants import GM_MOON, R_MOON
from lunanav.sim.simulator import Sim6DOF
```

# Rigid Body

```python
r = R_MOON + 15 # GEO
v = np.sqrt(GM_MOON/r) # circular velocity
print(r)
print(v)
```

```python
# Example with 

dt = .1
t0 = 0
n_steps = 86400 # About 2 days

state0 = np.array([
    r, 0, 0, 
    0,v,0, 
    1,0,0,0, 
    1,0,0])

sim = Sim6DOF(state0, t0, dt, n_steps)

sim.simulate()
```

```python
n = 80000
plt.plot(sim.X[:n,0], sim.X[:n,1])
# plt.axis('equal')
plt.xlim(r-10, r+10)
plt.ylim(-10, 10)
plt.show()
```
